# 顯示圖表的圖例

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
plt.plot(x, sinus, "r-o", label="sin(x)")
plt.plot(x, cosinus, "g--", label="cos(x)")
plt.legend()
plt.xlabel("徑度")
plt.ylabel("振幅")
plt.title("Sin和Cos三角函數的波型")
plt.show()
```

# Python程式準備使用TIOBE 2018年2月常用程式語言的使用率來繪製長條圖，首先建立3個串列和陣列，如下所示：


```python
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
index = np.arange(len(labels))
ratings = [5.168, 5.726, 14.988, 3.165, 11.857, 4.453]
plt.bar(index, ratings)
plt.xticks(index, labels)
plt.ylabel("使用率")
plt.title("程式語言的使用率") 
plt.show()
```

```python
plt.barh(index, ratings)
plt.yticks(index, labels)
plt.ylabel("使用率")
plt.title("程式語言的使用率")
plt.show()
```

```python
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
index = np.arange(len(labels)*2)
ratings = [5.168, 5.726, 14.988, 3.165, 11.857, 4.453]
change = [1.12, 0.3, -1.69, 0.29, 3.41, -0.45]
plt.bar(index[0::2], ratings, label="使用率")
plt.bar(index[1::2], change, label="增減值", color="r")
plt.legend()
plt.xticks(index[0::2], labels)
plt.ylabel("使用率")
plt.title("程式語言的使用率")
plt.show()
```

```python
labels = ["Python", "C++", "Java", "JS", "C", "C#"]
index = np.arange(len(labels)*3)
ratings = [5.168, 5.726, 14.988, 3.165, 11.857, 4.453]
changes = [1.12, 0.3, -1.69, 0.29, 3.41, -0.45]
randvals = np.random.rand(len(labels)) * 5
plt.bar(index[0::3], ratings, label="使用率")
plt.bar(index[1::3], changes, label="增減值", color="r")
plt.bar(index[2::3], randvals, label="隨機值", color="g")
plt.legend()
plt.xticks(index[1::3], labels, rotation=45)
plt.ylabel("使用率")
plt.title("程式語言的使用率")
plt.show()
```

# 繪製2張垂直排列的子圖表

```python
x = np.linspace(0.00001, 10, 50)
sinus = np.sin(x)
cosinus = np.cos(x)
tangent = np.tan(x)
cotangent = 1 / np.tan(x)
plt.subplot(2, 2, 1)
plt.plot(x, sinus, "r-o")
plt.subplot(2, 2, 2)
plt.plot(x, cosinus, "g--")
plt.subplot(2, 2, 3)
plt.plot(x, tangent, "b-o")
plt.subplot(2, 2, 4)
plt.plot(x, cotangent, "y--")
plt.show()
```

# 顯示共用x軸多軸圖表的圖例

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
sinhs = np.sinh(x)
fig, ax = plt.subplots()
ax.plot(x, sinus, "r-o", label="Sin(x)")
ax.set_xlabel("x", color="green")
ax.set_ylabel("Sin(x)", color="red")
ax.legend(loc="best")
ax2 = ax.twinx()
ax2.plot(x, sinhs, "g--", label="Sinh(x)")
ax2.set_ylabel("Cos(x)", color="blue")
ax2.legend(loc="best")
```

# 顯示共用x軸多軸圖表的圖例

```python
x = np.linspace(0, 10, 50)
sinus = np.sin(x)
sinhs = np.sinh(x)
fig, ax = plt.subplots()
lns1 = ax.plot(x, sinus, "r-o", label="Sin(x)")
ax.set_xlabel("x", color="green")
ax.set_ylabel("Sin(x)", color="red")
ax2 = ax.twinx()
lns2 = ax2.plot(x, sinhs, "g--", label="Sinh(x)")
ax2.set_ylabel("Sinh(x)", color="blue")
lns = lns1 + lns2
labs = [l.get_label() for l in lns]
ax.legend(lns, labs, loc="upper center")
```
