# Seaborn套件的基本使用

```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import math
```

## 繪製軸等級的圖表

```python
x = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,
     5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]
sinus = [math.sin(v) for v in x]
cosinus = [math.cos(v) for v in x]

sns.set()
fig, axes = plt.subplots(1,2, figsize=(6,4))
ax1 = sns.lineplot(x=x, y=sinus, ax=axes[0])
ax2 = sns.scatterplot(x=x, y=cosinus, ax=axes[1])
plt.show()
```

```python
x = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,
     5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]
sinus = [math.sin(v) for v in x]
cosinus = [math.cos(v) for v in x]
df = pd.DataFrame()
df["x"] = x
df["sin"]= sinus
df["cos"] = cosinus
print(df.head())
```

```python
df2 = pd.melt(df, id_vars=['x'], value_vars=['sin', 'cos'])
df2
```

```python
sns.set()
sns.relplot(x="x", y="value", kind="scatter", col="variable", data=df2)
plt.show()
```

## 更改Seaborn圖表的樣式
### 可用的參數值有：darkgrid（預設值）、whitegrid、dark、white和ticks

```python
x = [0,0.5,1,1.5,2,2.5,3,3.5,4,4.5,5,
     5.5,6,6.5,7,7.5,8,8.5,9,9.5,10]
sinus = [math.sin(v) for v in x]

sns.set_style("whitegrid")
sns.lineplot(x=x, y=sinus)
plt.show()
```

## 刪除上方和右方的軸線

```python
sns.set_style("whitegrid")
sns.lineplot(x=x, y=sinus)
sns.despine() # 刪除上方和右方的軸線
plt.show()
```

## 更改Seaborn佈景的樣式
### 在Seaborn的set_style()函數可以在第2個參數使用字典來更改佈景的細部樣式

```python
sns.set_style("darkgrid", {"axes.axisbelow": False})
sns.lineplot(x=x, y=sinus)
plt.show()
```

## 更改Seaborn圖表的外觀
### 除了使用Seaborn圖表的佈景外，Python程式一樣可以使用Matplotlib的函數來更改圖表的外觀

```python
plt.rcParams["axes.unicode_minus"] = False
sns.set_style("darkgrid", {"axes.axisbelow": False,
                       "font.sans-serif":['Microsoft JhengHei']})
sns.lineplot(x=x, y=sinus)
plt.title("Sinus三角函數的波型")
plt.xlim(-2, 12)
plt.ylim(-2, 2)
plt.xlabel("x")
plt.ylabel("sin(x)")
plt.show()
```

## 更改Seaborn圖表的尺寸
### Seaborn軸等級圖表函數可以使用Matplotlib函數更改圖表尺寸，圖形等級的圖表函數只能使用height和aspect參數

```python
sns.relplot(x="x", y="value", kind="scatter", col="variable",
            height=4, aspect=1.2, data=df2)
```

# 兩個數值資料的散佈圖

## 使用Seaborn繪製散佈圖

```python
df = sns.load_dataset("tips")
sns.set()
sns.relplot(x="total_bill", y="tip", data=df)
plt.show()
```

## 使用第三維的色調語意

```python
# 使用分類型欄位來指定不同點的色調（Hue）
sns.relplot(x="total_bill", y="tip", hue="smoker", data=df)
```

```python
# 使用不同標記樣式來顯示資料點
sns.relplot(x="total_bill", y="tip", hue="smoker",
            style="smoker", data=df)
```

# 時間趨勢的折線圖

## 一個時間點只有一筆觀察資料

```python
# 繪製Kobe Bryant生涯的平均每場得分趨勢
df = pd.read_csv("data/kobe_stats.csv")
df
```

```python
data = pd.DataFrame()
data["Season"] = pd.to_datetime(df["Season"])
data["PTS"] = df["PTS"]
sns.set()
sns.relplot(x="Season", y="PTS", data=data, kind="line")
plt.show()
```

# 資料集分佈情況的圖表

```python
df = sns.load_dataset("tips")
sns.set()
sns.histplot(df["total_bill"], 
                   kde=False)
plt.show()
```

## 迴歸圖表

```python

sns.lmplot(x="total_bill", y="tip", data=df)
```

### 繪製安斯庫姆四重奏資料集的迴歸線

```python
df = sns.load_dataset("anscombe")
sns.lmplot(x="x", y="y", col="dataset", hue="dataset", data=df,
           col_wrap=2, ci=None, height=4)
```

### 多項式迴歸模型

```python
sns.lmplot(x="x", y="y", data=df.query("dataset=='II'"), order=2)
```

### 使用殘差圖找出異常值

```python
sns.residplot(x="x", y="y", data=df.query("dataset=='III'"))
```
